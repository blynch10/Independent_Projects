```python
source5['letac'] = source5['Letter'] + ' ' + source5['Action']
source6 = source5.drop(['Letter', 'Action'], axis = 1)
source6
```

```python
source7 = source6.groupby(['letac'])['Number'].sum()
source7 = pd.DataFrame(source7).reset_index()
source7
```

```python
source8 = source7['letac'].str.split(' ', n=1, expand = True)
source8 = source8.rename(columns={0: 'Letter', 1: 'Action'})
source8['Number']=source7['Number']
source8 
```

```python
source9= source8[source8['Action'].str.contains('noise')]
source10 = source8[source8['Action'].str.contains('junk')]
source11 = pd.concat([source9, source10])
source12 = source11.sort_values(['Letter'])
source12
```

```python
source13= pd.DataFrame(source12).reset_index()
source14 = source13.drop(['index'], axis =1)
source14
```

```python
a=0
higher_value = []
higher_letter = []
while a<19:
    if source14['Letter'][a] == source14['Letter'][a+1]:
        if source14['Number'][a] > source14['Number'][a+1]:
            b = source14['Action'][a]
            c = source14['Letter'][a]
            higher_value.append(b)
            higher_letter.append(c)
            a=a+1
        else:
            b = source14['Action'][a+1]
            c = source14['Letter'][a+1]
            higher_value.append(b)
            higher_letter.append(c)
            a=a+1
    else:
        a=a+1
```

```python
higher_value = pd.DataFrame(higher_value)
higher_letter= pd.DataFrame(higher_letter)

higher_letter['Value'] = higher_value
#higher_table2['Letter']=higher_letter 
#higher_table2['Action'] = higher_value
#higher_table2
higher_letter
```

```python
higher_letter2 = higher_letter.drop(higher_letter[higher_letter.Value != 'junk'].index)
higher_letter2
```

```python
answer = higher_letter2.loc[:,0].apply(str)
answer2 = ",".join(answer)
answer2
```

```python
print('The sources that reported more junk than noise were: ' + answer2)
```
