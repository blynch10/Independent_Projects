```python
from skimage.io import imread, imshow
from skimage.filters import prewitt_h, prewitt_v
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image2 = imread('eyes.jpg', as_gray = True)
```

```python
edge_h = prewitt_h(image2)
edge_v = prewitt_v(image2)
imshow(edge_v)
```
